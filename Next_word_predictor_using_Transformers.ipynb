```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
```

```python

# Load GPT-2 model and tokenizer
model_name = "gpt2"  # You can also use "gpt2-medium", "gpt2-large" for better performance
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
```

```python
# Set the model to evaluation mode
model.eval()
```

```python

# Function to predict the next word in a sentence
def predict_next_word(input_text):
    # Encode the input text to tensor
    input_ids = tokenizer.encode(input_text, return_tensors="pt")

    # Predict the next token (word)
    with torch.no_grad():
        outputs = model(input_ids)
        predictions = outputs.logits

    # Get the predicted next token's ID (the last token)
    predicted_token_id = torch.argmax(predictions[0, -1, :]).item()

    # Decode the predicted token to a word
    predicted_word = tokenizer.decode([predicted_token_id])

    return predicted_word.strip()

# Chatbot function for predicting the next 5 words
def next_word_chat():
    print("Next Word Predictor: Type a sentence, and I'll predict the next 5 words one by one! (Type 'exit' to end)")

    while True:
        # Get user input
        user_input = input("You: ")

        if user_input.lower() == "exit":
            print("Next Word Predictor: Bye! Have a great day!")
            break

        # Predict the next 5 words one by one
        for i in range(10):
            # Get the predicted next word
            next_word = predict_next_word(user_input)

            # Print the word
            #print(f"Next Word Predictor: {next_word}")

            # Append the predicted word to the input text for the next prediction
            user_input += " " + next_word  # Concatenate the new word to the input
            print(user_input)

# Start the chat for next word prediction
next_word_chat()
```
